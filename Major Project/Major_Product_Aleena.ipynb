Data Cleaning

```python
# find all the columns that have null values
df.isnull().head()
```

```python
for i in df.columns:
  print(i)
  print(df[i].isnull().value_counts())
  print()
```

```python
# we can see that Evaporation & Sunshine columns have maximum number of null values. 
# similarly Cloud9am & Cloud3pm columns also have the next largest number of null values
# So we will drop those columns completely
df_noNulls = df.drop(['Evaporation','Sunshine','Cloud9am','Cloud3pm'],axis=1)
df_noNulls.head()
```

```python
print('Columns that have null values: ')
hasNull = []
for i in df_noNulls.columns:
  for j in df_noNulls[i].isnull().values:
    if j==True:
      hasNull.append(i)
      break
print(hasNull)
```

```python
# remove all the rows containing null values
df_noNulls.dropna(inplace=True, subset=hasNull)
df_noNulls.head()
```

```python
df_noNulls.isnull().head()
```

```python
print('shape: ',df_noNulls.shape)
print('size: ',df_noNulls.size)
```

```python
# Slice dataframe to show only locations and their amount of rainfall
locRain = df_noNulls[['Location','Rainfall']]
locRain
```

**Linear Regression**

> Find a relation between the wind speeds at 9am and 3pm as well as the wind gust speed



```python
wind_speeds = df[['WindGustSpeed','WindSpeed9am','WindSpeed3pm']].dropna()
wind_speeds.head()
```

```python
plt.scatter(wind_speeds['WindSpeed9am'],wind_speeds['WindSpeed3pm'])
plt.title('Wind Speed at 9am v/s Wind Speed at 3pm')
plt.xlabel('Wind Speed at 9am')
plt.ylabel('Wind Speed at 3pm')
```

```python
plt.scatter(wind_speeds['WindSpeed9am'],wind_speeds['WindGustSpeed'])
plt.title('Wind Speed at 9am v/s Wind Gust Speed')
plt.xlabel('Wind Speed at 9am')
plt.ylabel('Wind Gust Speed')
```

```python
plt.scatter(wind_speeds['WindSpeed3pm'],wind_speeds['WindGustSpeed'])
plt.title('Wind Speed at 3pm v/s Wind Gust Speed')
plt.xlabel('Wind Speed at 3pm')
plt.ylabel('Wind Gust Speed')
```

```python
X = wind_speeds.drop('WindGustSpeed',axis=1)
Y = wind_speeds['WindGustSpeed']

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.30)
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
```

```python
# Fit the train data to the model
lm.fit(X_train, Y_train)
```

```python
pred = lm.predict(X_test)
```

```python
plt.scatter(Y_test,pred)
plt.title('Comparing actual and predicted outputs')
plt.xlabel('Actual Output')
plt.ylabel('Predicted Output')
```

**Logistic Regression**

> Use the dataset without Evaporation, Sunshine, Cloud9am, Cloud3pm columns


> Predict whether it will rain tomorrow

```python
# remove columns that have data type of object
final_data = df_noNulls.drop(['Date','Location','WindGustDir','WindDir9am','WindDir3pm'],axis=1)
final_data.head()
```

```python
# change the RainToday and RainTomorrow column values to 0 and 1 instead of No and Yes
def change_val(val):
  if val=='Yes':
    return 1
  else:
    return 0
final_data['RainToday'] = final_data['RainToday'].apply(change_val)
final_data['RainTomorrow'] = final_data['RainTomorrow'].apply(change_val)
```

```python
final_data.info()
```

```python
X = final_data.drop(['RainTomorrow'],axis=1)
Y = final_data['RainTomorrow']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.30)
```

```python
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression()
```

```python
# fit the train data into the logistic regression model
log_model.fit(X_train, Y_train)
```

```python
pred = log_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print("Confusion Matrix: ")
print(confusion_matrix(Y_test,pred))
print()

print("Classification report: ")
print(classification_report(Y_test,pred))
```
